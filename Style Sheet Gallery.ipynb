```python
import matplotlib.pyplot as plt
import geopandas as gpd
import seaborn as sns
import pandas as pd
import random
import matplotlib.colors as colors
random.seed(43)
plt.style.use("ALLFED.mplstyle")
```

```python
x = [random.random() for i in range(100)]
y = [random.random() for i in range(100)]
z = [random.random() for i in range(100)]
```

```python
plt.plot(x)
ax = plt.gca()
ax.set_title("Font");
ax.set_ylabel("Y-Axis")
ax.set_xlabel("X-Axis");
```

```python
plt.scatter(x,y);
ax = plt.gca()
ax.set_ylabel("Y-Axis")
ax.set_xlabel("X-Axis");
```

```python
plt.boxplot(x)
# Remove the grid in certain plots
ax = plt.gca()
ax.xaxis.grid(False)
ax.set_ylabel("Y-Axis")
ax.set_xlabel("Category")
```

```python
plt.hist(y);
ax = plt.gca()
ax.set_ylabel("Y-Axis")
ax.set_xlabel("X-Axis");
```

```python
sns.kdeplot(x=x,y=y,fill=True);
ax = plt.gca()
ax.set_ylabel("Y-Axis")
ax.set_xlabel("X-Axis");
```

```python
path_to_data = gpd.datasets.get_path("nybb")
gdf = gpd.read_file(path_to_data)
```

```python
gdf.plot();
ax = plt.gca()
ax.set_ylabel("Y-Axis")
ax.set_xlabel("X-Axis");
```

```python
# Create a custom colormap with the specified colors
# geopandas only takes colormaps, so we have to define it here
colors_cmap =[ '#3A913F', '#363631', '#DC582A', '#F0B323', '#3D87CB']
cmap = colors.LinearSegmentedColormap.from_list("my_colormap", colors_cmap)

ax = gdf.plot(column="BoroName", legend=False, cmap=cmap);
ax.set_title("Categorical Color");
ax = plt.gca()
ax.set_ylabel("Y-Axis")
ax.set_xlabel("X-Axis");
```

```python
ax = gdf.plot(column="Shape_Leng", legend=True, legend_kwds={"label":"title"});
ax.set_title("Continuous Color");
ax = plt.gca()
ax.set_ylabel("Y-Axis")
ax.set_xlabel("X-Axis");
```

```python
# Load the example diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Draw a scatter plot while assigning point colors and sizes to different
# variables in the dataset
f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.despine(f, left=True, bottom=True)
clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.scatterplot(x="carat", y="price",
                hue="clarity", size="depth",
                palette="viridis",
                hue_order=clarity_ranking,
                sizes=(1, 8), linewidth=0,
                data=diamonds, ax=ax)
```
