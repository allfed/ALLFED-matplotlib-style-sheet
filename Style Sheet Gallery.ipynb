```python
import matplotlib.pyplot as plt
import geopandas as gpd
import seaborn as sns
import random
random.seed(43)
plt.style.use("ALLFED.mplstyle")
```

```python
x = [random.random() for i in range(20)]
y = [random.random() for i in range(20)]
```

```python
plt.plot(x)
ax = plt.gca()
ax.set_title("Serif Font");
```

```python
plt.scatter(x,y);
```

```python
plt.boxplot(x)
# Remove the grid in certain plots
ax = plt.gca()
ax.xaxis.grid(False)
```

```python
plt.hist(y);
```

```python
sns.kdeplot(x=x,y=y,fill=True);
```

```python
path_to_data = gpd.datasets.get_path("nybb")
gdf = gpd.read_file(path_to_data)
```

```python
gdf.plot();
```

```python
ax = gdf.plot(column="BoroName", legend=True);
ax.set_title("Categorical Color");
```

```python
ax = gdf.plot(column="Shape_Leng", legend=True, legend_kwds={"label":"title"});
ax.set_title("Continuous Color");
```
