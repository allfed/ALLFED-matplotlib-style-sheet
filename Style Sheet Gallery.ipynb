```python
import matplotlib.pyplot as plt
import seaborn as sns
import random
random.seed(43)
plt.style.use("ALLFED.mplstyle")
```

```python
x = [random.random() for i in range(20)]
y = [random.random() for i in range(20)]
```

```python
plt.plot(x)
ax = plt.gca()
ax.set_title("Serif Font")
```

```python
plt.scatter(x,y)
```

```python
plt.boxplot(x)
# Remove the grid in certain plots
ax = plt.gca()
ax.xaxis.grid(False)
```

```python
plt.hist(y)
```

```python
sns.kdeplot(x=x,y=y,fill=True)
```
